### Using Our Shiny New Uploader
Now that you've got your repository ready, here's how to use our streamlined uploader:

1. **Repository Details** ğŸ“š
   - Fill in your Organization/Username
   - Add your Repository name (the one you just created!)

2. **File Selection** ğŸ—‚ï¸
   - Pick your file type from the dropdown (SafeTensors, PyTorch, etc.)
   - Choose how to sort your files (by name or date)
   - Select the files you want to upload from the list

3. **Upload Settings** ğŸš€
   - Want to create a Pull Request? Just tick the checkbox!
   - Hit the "Upload" button and watch your files take flight!

The uploader will keep you posted with friendly messages about how things are going. If anything goes sideways, it'll let you know what happened and how to fix it! 

Remember: Your repository needs to exist before you start uploading - think of it like having your destination ready before starting your journey! âœ¨

Need to upload more files later? No problem! Just repeat these steps whenever you need to send more files to your repository. Easy peasy! ğŸ¯


```python
import glob
import os
from pathlib import Path
from huggingface_hub import HfApi
from ipywidgets import Text, Dropdown, Button, SelectMultiple, VBox, HBox, Output, Layout, Checkbox, HTML, Textarea
from IPython.display import display, clear_output
import time

class HuggingFaceUploader:
    def __init__(self):
        self.api = HfApi()
        self.file_types = [
            # AI Model Files ğŸ¤–
            ('SafeTensors', 'safetensors'),
            ('PyTorch Models', 'pt'),
            ('PyTorch Legacy', 'pth'),
            ('ONNX Models', 'onnx'),
            ('TensorFlow Models', 'pb'),
            ('Keras Models', 'h5'),
            
            # Checkpoint Files ğŸ¯
            ('Checkpoints', 'ckpt'),
            ('Binary Files', 'bin'),
            
            # Config & Data Files ğŸ“
            ('JSON Files', 'json'),
            ('YAML Files', 'yaml'),
            ('YAML Alt', 'yml'),
            ('Text Files', 'txt'),
            ('CSV Files', 'csv'),
            ('Pickle Files', 'pkl'),
            
            # Image Files ğŸ¨
            ('PNG Images', 'png'),
            ('JPEG Images', 'jpg'),
            ('JPEG Alt', 'jpeg'),
            ('WebP Images', 'webp'),
            ('GIF Images', 'gif'),
            
            # Archive Files ğŸ“¦
            ('ZIP Archives', 'zip'),
            ('TAR Files', 'tar'),
            ('GZ Archives', 'gz')
        ]
        self._create_widgets()
        self._bind_events()
        
    def _create_widgets(self):
        # Repository info section
        self.repo_info = HTML(value="<b>ğŸ“š Repository Details</b>")
        self.org_name = Text(
            placeholder='Organization or Username',
            description='Owner:',
            style={'description_width': 'initial'}
        )
        self.repo_name = Text(
            placeholder='Repository Name',
            description='Repo:',
            style={'description_width': 'initial'}
        )
        
        # File handling section
        self.file_section = HTML(value="<b>ğŸ—‚ï¸ File Selection</b>")
        self.file_type = Dropdown(
            options=self.file_types,
            value='safetensors',
            description='File Type:',
            style={'description_width': 'initial'}
        )
        self.sort_by = Dropdown(
            options=['name', 'date'],
            value='name',
            description='Sort By:'
        )
        
        # Custom commit message
        self.commit_section = HTML(value="<b>ğŸ’­ Commit Details</b>")
        self.commit_msg = Textarea(
            value="Upload with Earth & Dusk Huggingface ğŸ¤— Backup",
            placeholder='Enter your commit message',
            description='Message:',
            layout=Layout(width='400px', height='60px')
        )
        
        # Upload options section
        self.upload_section = HTML(value="<b>ğŸš€ Upload Settings</b>")
        self.create_pr = Checkbox(
            value=False,
            description='Create Pull Request',
            indent=False,
            tooltip='When checked, creates a Pull Request instead of direct upload'
        )
        self.clear_after = Checkbox(
            value=True,
            description='Clear output after upload',
            indent=False,
            tooltip='Clears output area after successful upload'
        )
        
        # Progress tracking
        self.progress_text = HTML(value="")
        
        # Action buttons with style
        self.update_btn = Button(
            description='ğŸ”„ Update Files',
            button_style='info',
            tooltip='Refresh file list'
        )
        self.upload_btn = Button(
            description='â¬†ï¸ Upload',
            button_style='success',
            tooltip='Start upload process'
        )
        self.clear_btn = Button(
            description='ğŸ§¹ Clear Output',
            button_style='warning',
            tooltip='Clear output area'
        )
        
        # File selector and output
        self.ckpt_picker = SelectMultiple(
            options=[], 
            layout=Layout(width="600px", height="200px")
        )
        self.out = Output()
        
    def _bind_events(self):
        self.update_btn.on_click(self._update_files)
        self.upload_btn.on_click(self._upload_ckpts)
        self.clear_btn.on_click(lambda _: self.out.clear_output())
        self.file_type.observe(self._update_files, names='value')
        
    def _update_files(self, _):
        file_extension = self.file_type.value
        all_ckpts = sorted(
            glob.glob(f"*.{file_extension}"),
            key=os.path.getmtime if self.sort_by.value == 'date' else str
        )
        self.ckpt_picker.options = all_ckpts
        
        with self.out:
            print(f"âœ¨ Found {len(all_ckpts)} {file_extension} files")
            
    def _format_size(self, size):
        for unit in ['B', 'KB', 'MB', 'GB']:
            if size < 1024:
                return f"{size:.2f} {unit}"
            size /= 1024
        return f"{size:.2f} TB"
            
    def _upload_ckpts(self, _):
        if not self.org_name.value or not self.repo_name.value:
            with self.out:
                print("â— Please fill in both Organization/Username and Repository name")
                return
                
        repo_id = f"{self.org_name.value}/{self.repo_name.value}"
        total_files = len(self.ckpt_picker.value)
        
        with self.out:
            if not self.ckpt_picker.value:
                print("ğŸ“ Nothing selected for upload. Please select files from the list.")
                return
            
            print(f"ğŸ¯ Starting upload to: huggingface.co/{repo_id}")
            
            for idx, ckpt in enumerate(self.ckpt_picker.value, 1):
                file_size = os.path.getsize(ckpt)
                print(f"\nğŸ“¦ File {idx}/{total_files}: {ckpt} ({self._format_size(file_size)})")
                
                try:
                    start_time = time.time()
                    response = self.api.upload_file(
                        path_or_fileobj=ckpt,
                        path_in_repo=os.path.basename(ckpt),
                        repo_id=repo_id,
                        repo_type=None,
                        create_pr=self.create_pr.value,
                        commit_message=self.commit_msg.value
                    )
                    duration = time.time() - start_time
                    print(f"âœ… Upload completed in {duration:.1f} seconds")
                    display(response)
                except Exception as e:
                    print(f"âŒ Error uploading {ckpt}: {str(e)}")
                    continue
            
            print("\nâœ¨ All uploads completed! âœ¨")
            if self.create_pr.value:
                print("ğŸ‰ Check your repository for the new Pull Request!")
            else:
                print("ğŸ‰ Files have been uploaded directly to your repository!")
                
            if self.clear_after.value:
                time.sleep(3)  # Give user time to read messages
                self.out.clear_output()
            
    def display(self):
        box = VBox([
            self.repo_info,
            HBox([self.org_name, self.repo_name]),
            self.file_section,
            HBox([self.file_type, self.sort_by]),
            self.commit_section,
            self.commit_msg,
            self.upload_section,
            HBox([self.create_pr, self.clear_after]),
            self.update_btn,
            self.ckpt_picker,
            HBox([self.upload_btn, self.clear_btn]),
            self.progress_text,
            self.out
        ])
        display(box)

# Create and display the uploader - just one line to rule them all! âœ¨
uploader = HuggingFaceUploader()
uploader.display()
```
